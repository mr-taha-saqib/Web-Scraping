```python
import csv
from bs4 import BeautifulSoup
import requests

url = 'https://www.dawnnews.tv/tech'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
data = []

for article in soup.find_all('article', class_='story'):
    story_title = article.find('h2', class_='story__title')
    story_link = article.find('a', class_='story__link')
    media_item = article.find('div', class_='media__item')
    story_excerpt = article.find('div', class_='story__excerpt')
    story_time = article.find('span', class_='story__time')

    if all([story_title, story_link, media_item, story_excerpt, story_time]):
        data.append([
            story_title.text.strip(),
            story_link['href'].strip(),
            media_item.text.strip(),
            story_excerpt.text.strip(),
            story_time.text.strip()
        ])
```

```python

csv_filename = 'scraped_data_Q2.csv'
with open(csv_filename, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Title', 'Link', 'Media Item', 'Excerpt', 'Time'])
    writer.writerows(data)
print(f'Data has been saved to {csv_filename}')
```
